## 将逻辑条件作为数组操作 ：np.where
np.where()要传入3个参数，第一个是包含布尔值的数组，第二第三个可以是标量也可以是数组。

```python
import numpy as np
arr = np.random.randn(1,4)
arr_1 = [1,2,3,4]
```

```python
arr
```

```python
arr_bool = [True,True,False,False]
```

```python
np.where(arr_bool,arr,arr_1)
```

使用np.where，可以将标量和数组结合起来。例如，我可用常数2替换arr中所有正的值：

```python
np.where(arr>0,2,arr)
```

## 随机漫步

从0开始，步长1和－1出现的概率相等。

```python
import random
position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(walk[:100])
```
